#Aliasing and Nyquist



**Nyquist's theorem:** A theorem, developed by H. Nyquist, which states that an analog signal waveform may be uniquely reconstructed, without error, from samples taken at equal time intervals. The sampling rate must be equal to, or greater than, twice the highest frequency component in the analog signal.
<hr>

**Aliasing** refers to the distortion or artifact that results when a signal reconstructed from samples is different from the original continuous signal.
<hr>
The standard ECG is by convention limited to 0.05-150 Hz
<hr>
in this tutorial, the given ECG signal is sampled by 500 Hz which means it can be constructed with no distortion   


let's discover the aliasing on the given ECG signal

```python
def my_max(arr):                #implement a function that returns the maximum value then compare your result with numpy library
    max_value = arr[0]

    for value in arr:
        if value > max_value:
            max_value = value

    return max_value

my_max(first2Seconds)
```

```python
#implement a function that returns the minimum value then compare your result with numpy library
def my_min(arr):
    min_value = arr[0]
    for value in arr:
        if value < min_value:
            min_value = value
    return min_value

my_min(first2Seconds)
```

```python
print("Numpy min:", np.min(first2Seconds))
print("My function min:", my_min(first2Seconds))
```

```python
#find the mean using numpy library
np.mean(first2Seconds)
```

```python
#implement a function that returns the mean then compare your result with numpy library
def my_mean(arr):
    total = 0
    for value in arr:
        total += value
    return total / len(arr)

my_mean(first2Seconds)
```

```python
print("Numpy mean:", np.mean(first2Seconds))
print("My function mean:", my_mean(first2Seconds))
```

```python
#Calculate the median using numpy library
np.median(first2Seconds)
```

```python
#implement a function that returns the median then compare your result with numpy library
def my_median(arr):
    sorted_arr = sorted(arr)
    n = len(sorted_arr)
    mid = n // 2

    if n % 2 == 0:
        return (sorted_arr[mid-1] + sorted_arr[mid]) / 2
    else:
        return sorted_arr[mid]

my_median(first2Seconds)
```

```python
print("Numpy median:", np.median(first2Seconds))
print("My function median:", my_median(first2Seconds))
```

```python
import heartpy as hp
```

```python
ecg_signal = data["'ECG I'"]
ecg_signal = pd.to_numeric(ecg_signal, errors='coerce')
ecg_signal = ecg_signal.dropna().to_numpy()
```

```python
fs = 500
```

```python
filtered_signal = hp.filter_signal(ecg_signal, cutoff=5, sample_rate=fs, order=3, filtertype='lowpass')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(ecg_signal)
plt.title("Original ECG Signal")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()
```

```python
fs = 500

filtered_signal = hp.filter_signal(
    ecg_signal,
    cutoff = 5,
    sample_rate = fs,
    order = 3,
    filtertype = 'lowpass'
)
```

```python
plt.figure(figsize=(12,4))
plt.plot(filtered_signal)
plt.title("Filtered ECG Signal (Lowpass)")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()
```
